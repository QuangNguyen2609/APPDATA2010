# Section 1 - Blodd Glucose Data
+ The data of this section is about blood glucose readings for 500 persons without diabetes. Additionally, normal values are below 126 mg/dL and over 60 mg/dL
+ Use python to sample this data for 500 people

```python
import random

glucose_data = [round(random.uniform(61, 125), 2) for _ in range(500)] 
for i in range(500):
    print(f"blood glucose of patient {i+1}:", glucose_data[i], "mg/dL")
```

# Section 2: Unit and data point chooser
+ In Australia, blood glucose levels are reported in mmol/L instead of mg/dL. The conversion is : mmol/L. The conversion is mmol/L = (mg/dL) / 18
+ Use python to convert from mg/dL to mmol/L from user input

```python
unit = input("Enter the preferred unit (mmol/L or mg/dL): ")
print("The preferred unit is:", unit)
if unit == "mmol/L":
    glucose_data = [round(random.uniform(61, 125) / 18, 2) for _ in range(500)]
    for i in range(500):
        print(f"blood glucose of patient {i+1}:", glucose_data[i], "mmol/L")
else:
    glucose_data = [round(random.uniform(61, 125), 2) for _ in range(500)]
    for i in range(500):
        print(f"blood glucose of patient {i+1}:", glucose_data[i], "mg/dL")
```

# Section 3: Giving more choices
+ This is the extension from Section 2 that this part allows user to input number of data points that they want to generate and also the preferred unit 

```python
number_of_patients = int(input("Enter the number of patients: "))
print("The number of patients is:", number_of_patients)
unit = input("Enter the preferred unit (mmol/L or mg/dL): ")
print("The preferred unit is:", unit)
if unit == "mmol/L":
    glucose_data = [round(random.uniform(61, 125) / 18, 2) for _ in range(number_of_patients)]
    for i in range(number_of_patients):
        print(f"blood glucose level of patient {i+1}:", glucose_data[i], "mmol/L")
else:
    glucose_data = [round(random.uniform(61, 125), 2) for _ in range(number_of_patients)]
    for i in range(number_of_patients):
        print(f"blood glucose level of patient {i+1}:", glucose_data[i], "mg/dL")
```
